```python
##types of arguments
#positional arguments
#keyword arguments
#default arguments
#variable lengths of positional arguments
#variable lengths of keyword arguments


#positional arguments
# subtraction of two numbers using functions
def sub (x,y):
    res = x-y
    return res
a = int(input())
b = int(input())   ##a,b =10,20
print(sub(a,b))
print(sub(b,a))
```

```python
##keyword arguments
##multiplication os two numbers
a = int(input("enter a number 1:"))
b= int(input("enter a number 2:"))
def multiplication(n1,n2):
    res = n1  * n2
    return res
print (multiplication (n1=a,n2=b))
print(multiplication(n2=b,n1=a))

```

```python
###checking even or odd 
def even_or_odd(num):
    if num % 2:
        return "even"
    else:
        return "odd"
a = int(input("enter a number"))
res=even_or_odd(a)
print(res)
```
